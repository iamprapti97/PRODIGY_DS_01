**VISUAL REPRESENTATION (BAR DIAGRAM)**

```python
age_counts = dataset['age'].value_counts()
```

```python
print(type(age_counts))
```

```python
colors = np.linspace(0.2, 0.8, len(age_counts))
colors = plt.cm.Reds(colors)

plt.bar(age_counts.index, age_counts, color=colors)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age distribution for smokers")
plt.show()
```
